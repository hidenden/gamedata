```python
def chart_reached_unit(n:int) -> pd.DataFrame:
    df = hs.load_hard_sales()
    d2 = hs.extract_week_reached_units(df, n).sort_values('delta_week', ascending=True).head(10)
    d3 = d2[['report_date', 'hw', 'delta_week', 'sum_units']].copy()
    d3 = d3.copy().assign(delta_week=lambda x: x['delta_week'] + 1)
    d3.reset_index(drop=True, inplace=True)
    d3.index += 1
    return d3
    
```
