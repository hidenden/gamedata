```python
df_all = g.load_hard_sales()
```

```python
def add_weeek_number(df: pl.DataFrame) -> pl.DataFrame:
    if "delta_week" in df.columns:
        df = df.with_columns(
            (pl.col("delta_week") + 1).alias("week_number")
        )
    return df
```

```python
hard_n = 4800000
df2 = (
    g.extract_week_reached_units(df_all, hard_n).
    sort("delta_week", descending=False).
    head(10).
    select(["full_name", "report_date", "hw", "delta_week", "sum_units"])
)
df3 = (
    add_weeek_number(df2).
    select(["full_name", "report_date", "week_number"])
)
df4 = g.rename_columns(df3)
st = g.style(df4)
display(st)
```

```python
def chart_reached_unit(n:int) -> pl.DataFrame:
    df = g.load_hard_sales()
    df = (
        g.extract_week_reached_units(df, n).
        sort("delta_week", descending=False).
        head(10)
    )
    df = (add_weeek_number(df).
          select(["hw", "report_date", "week_number", "sum_units"]))
    return df
```

```python
def chart_reached_unit2(n:int) -> pl.DataFrame:
    df = g.load_hard_sales()
    d2 = g.extract_week_reached_units(df, n).sort_values('delta_week', ascending=True).head(10)
    d3 = d2[['report_date', 'hw', 'delta_week', 'sum_units']].copy()
    d3 = d3.copy().assign(delta_week=lambda x: x['delta_week'] + 1)
    d3.reset_index(drop=True, inplace=True)
    d3.index += 1
    return d3
    
```
