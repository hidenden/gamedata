```python
import matplotlib
```

```python
VERBOSE = False
PUBLISH = False

if not PUBLISH:
    %matplotlib widget
else:
    %matplotlib inline
```

```python
backend = matplotlib.get_backend()

print(f"現在のMatplotlibバックエンド: {backend}")

if 'backend_inline' in backend:
    print("現在は 'inline' モードです（静止画表示）")
elif 'ipympl' in backend or 'nbagg' in backend or 'widget' in backend:
    print("現在は 'widget' モードです（インタラクティブ表示）")
else:
    print(f"その他のバックエンドが有効です: {backend}")
```
