```python
### 週販レポート2025年版

# 標準ライブラリ
from datetime import datetime, timedelta

# サードパーティライブラリ
import polars as pl
import great_tables as gt
import matplotlib.pyplot as plt
import japanize_matplotlib
from matplotlib.ticker import ScalarFormatter
from IPython.display import Markdown, display

import gamedata as g

pl.Config.set_tbl_rows(40)
pl.Config.set_tbl_cols(-1)
```

```python
base_df = g.load_hard_sales()
base_df
```

```python
dated_df =(base_df
 .filter(pl.col("report_date") >= datetime(2026, 1, 1))
 .filter(pl.col("report_date") < datetime(2026, 2, 1))
 .sort("report_date")
)

display_df = dated_df.select([
    pl.col("report_date"),
    pl.col("full_name"),
    pl.col("units"),
    pl.col("sum_units")]).sort(["report_date", "units"], descending=[False, True])

display_df
                    

                
            
    
```

```python
gt1 = (
    gt.GT(
        display_df,
        groupname_col = "report_date",
        auto_align=True,
        locale="ja_JP",
    )
    .tab_options(
        table_font_size="1em",
        table_layout="auto",
        table_background_color="#FFFFFF00",
        table_font_color="#FFFFFF"
    )
    .fmt_integer(columns="units")
)


gt1
```
