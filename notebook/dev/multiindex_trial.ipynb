```python
# 標準ライブラリ
from datetime import datetime, timedelta

# サードパーティライブラリ
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# プロジェクト内モジュール
from gamedata import hard_sales as hs

df = hs.load_hard_sales()
```

```python
mdf1 = df.set_index(['year', 'month', 'week'])
mdf1.sort_values(by=['year', 'month', 'week', 'units'], ascending=[True, True, True, False], inplace=True)
mdf1
```

```python
hs.pivot_sales(mdf1)
```

```python
# 2025年 6, 7月のデータを取得
m67df = mdf1.loc[(2025, [5, 6, 7]), :]
m67df[["full_name", "units", "sum_units"]]
```

```python
mp1 = m67df.pivot_table(
    columns=["full_name"],
    index=["year", "month", "week"],
    values=["units", "sum_units"],
    aggfunc={"units": "sum", "sum_units": "sum"}
)

mp1 = mp1.swaplevel(axis=1).sort_index(axis=1)
mp1 = mp1.reindex(['units', 'sum_units'], axis=1, level=1)
```
