# 欠損値の確認を行う

## 欠損値の定義

HWの売り上げデータなので発売前はもちろんデータは無くなる。
発売後も時間が経過すれば集計対象から外れるのでデータは無くなる。
これは正常である。
問題になるのは期間途中のデータが無いものである。
これをどう抽出すべきか？

## 作戦



```python
import pandas as pd
import numpy as np
import gamedata.hard_sales as hs
from typing import List

df = hs.load_hard_sales()
pdf = hs.pivot_sales(df)
pdf
```

```python
hw_df = pdf["3DS"]
hw_df = hw_df.dropna()

data_diffs = hw_df.index.to_series().diff().dt.days
data_diffs = data_diffs.iloc[1:]
irregular = data_diffs[data_diffs != 7]

if irregular.empty:
    print("全てのインデックス間隔が7日です。")
else:
    print("7日でない間隔が存在します:")
    display(irregular)
```

```python
def check_irregular_intervals(df, hw_list:List[str]) -> None:
    for hw in hw_list:
        print(f"Checking irregular intervals for {hw}...")
        hw_df = df[hw].dropna()
        data_diffs = hw_df.index.to_series().diff().dt.days
        data_diffs = data_diffs.iloc[1:]
        irregular = data_diffs[data_diffs != 7]

        if irregular.empty:
            print(f"全てのインデックス間隔が7日です: {hw}")
        else:
            print(f"{hw}で7日でない間隔が存在します:")
            display(irregular)

```

```python
check_irregular_intervals(pdf, hs.get_hw_names(df))
print("全てのハードウェアのチェックが完了しました。"    )
```
