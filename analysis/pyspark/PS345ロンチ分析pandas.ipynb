```python
import pandas as pd
import numpy as np
import datetime
import re
import matplotlib.pyplot as plt
import matplotlib as mpl
import japanize_matplotlib
```

```python
hw_cumsum = pd.read_pickle("../database/pkl/weekly_hard_cumsum.pkl")
```

```python
launch_sum = pd.read_pickle("../database/pkl/launch_hard_cumsum.pkl")
```

```python
ps_launch = launch_sum[['PS4', 'PS5', 'PS3']].dropna(how="all")
ps_launch.interpolate(method="linear", limit=4).plot(figsize=(20, 10), grid=True)
```

```python
ps_launch["PS5"].dropna()
```

```python
'''
plot

xticks sequence
Values to use for the xticks.
yticks sequence
Values to use for the yticks.
xlim 2-tuple/list
Set the x limits of the current axes.
ylim 2-tuple/list
Set the y limits of the current axes.
'''

```

```python
xticks_list=[]
for y in range(0, 3500, 500):
    xticks_list.append(y)
xticks_list.append(936)
xticks_list.append(1027)
xticks_list = sorted(xticks_list)
```

```python
ps_launch.interpolate(method="linear", limit=1).plot(figsize=(18, 8),
                                                     grid=True, 
                                                     ylim=[0, 11000000],
                                                     xticks=xticks_list
                                                    )
plt.savefig('ps_lifetime.png')
```

```python
plt.close('all')
```
